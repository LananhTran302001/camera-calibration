### 1. Find corners

```python
image_1 = cv2.imread("assets/data_2.jpg")
gray = cv2.cvtColor(image_1, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(15, 9))
plt.imshow(gray, cmap="gray")
```

```python
ret, corners = cv2.findChessboardCorners(gray, (13,9), None)
if ret:
    print(f"Found {len(corners)} corners")
    cv2.drawChessboardCorners(image_1, (13, 9), corners, ret)
    plt.figure(figsize=(15, 9))
    plt.imshow(image_1)
```
